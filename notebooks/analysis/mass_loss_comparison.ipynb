```python
out_path = os.path.join(fig_dir,'mass_loss_obs_and_models.png')

plt.close()
fig,axes = plt.subplots(nrows=2,figsize=(15,20),gridspec_kw={'height_ratios': [9, 1]})
ax0,ax1 = axes.flatten()

data = mlc[['Liu 2015 (non steady)','Depoorter 2013','Rignot 2013 (IS flux)','This study',
            'Naughten 2018 (MetROMS)','Naughten 2018 (FESOM HR)','Mathiot 2017',
            'Timmermann 2012','Hellmer 2004']]

errors = mlc[['Liu 2015 err (non steady)','Depoorter 2013 err','Rignot 2013 err']]
errors.columns=['Liu 2015 (non steady)','Depoorter 2013','Rignot 2013 (IS flux)']


data.iloc[3:39:3].plot(kind='barh',xerr=errors.iloc[3:39:3],ax=ax0,width=0.8,fontsize=24,alpha=0.7,edgecolor='k',
                       color=['coral','red','darkred','lavender','darkgreen','palegreen','yellowgreen','limegreen','forestgreen'])

data = mlc[['Liu 2015 (non steady)','Depoorter 2013','Rignot 2013 (IS flux)','This study (all ice)',
            'Naughten 2018 (MetROMS)','Naughten 2018 (FESOM HR)','Mathiot 2017',
            'Timmermann 2012','Hellmer 2004']]

data.iloc[0:1].plot(kind='barh',xerr=errors.iloc[0:1],ax=ax1,width=0.8,fontsize=24,alpha=0.7,edgecolor='k',
                       color=['coral','red','darkred','lavender','darkgreen','palegreen','yellowgreen','limegreen','forestgreen'])

ax0.legend('')

handles, labels = ax0.get_legend_handles_labels()
labels[2] = 'Rignot 2013'
ax1.legend(handles[::-1], labels[::-1], loc='lower center', bbox_to_anchor=(0.3, -2),ncol=3,fontsize=24)
#ax.set_title('Ice shelf basal mass loss from models',fontsize=24)
ax1.set_xlabel('Mass loss in Gt/yr',fontsize=24)
ax0.grid(axis='x')
ax1.grid(axis='x')
ax0.set_xlim(0,350)
fig.tight_layout()
plt.savefig(out_path,dpi=200,format='png',bbox_to_anchor=True,bbox_inches='tight')
plt.show()
```
